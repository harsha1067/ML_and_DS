Q3: Create a short dataframe that shows the top 3 *liberal* arts and *state* schools that produce the highest (mid-career) earning graduates.

Show the *School Name* and *Mid-Career Median Salary* columns from each dataset, side by side, i.e. <ins>horizontally</ins>.

Nest the column labels within 'Liberal Arts' and 'State' labels.

```python
import pandas as pd

# Assuming 'liberal_url' and 'state_url' are URLs to your data files
# Replace these with your actual URLs

# Load liberal arts and state dataframes
liberal = pd.read_csv(liberal_url)
state = pd.read_csv(state_url)

# Convert 'Mid-Career Median Salary' column to numeric after cleaning
liberal['Mid-Career Median Salary'] = pd.to_numeric(liberal['Mid-Career Median Salary'].replace('[\$,]', '', regex=True))
state['Mid-Career Median Salary'] = pd.to_numeric(state['Mid-Career Median Salary'].replace('[\$,]', '', regex=True))

# Top 3 liberal arts schools with highest mid-career median salaries
top_liberals = liberal.nlargest(3, 'Mid-Career Median Salary')[['School Name', 'Mid-Career Median Salary']]

# Top 3 state schools with highest mid-career median salaries
top_states = state.nlargest(3, 'Mid-Career Median Salary')[['School Name', 'Mid-Career Median Salary']]

# Concatenate the dataframes horizontally with nested column labels
result_df = pd.concat({
    'Liberal Arts': top_liberals.set_index('School Name'),
    'State': top_states.set_index('School Name')
}, axis=1)

# Display the result
print("Top 3 Liberal Arts and State Schools with Highest Mid-Career Median Salaries:")
print(result_df)

```

Q13: Starting with our main dataframe (*sat*), create a pivot table that summarizes the <ins>total</ins> student enrollment (Student Enrollment) across all 5 boroughs (*Borough*).

Which boroughs have the highest and lowest high school student bodies?

```python
#Q13
# Pivot table to summarize total student enrollment across all 5 boroughs
pivot_table_enrollment = sat.pivot_table(values='Student Enrollment', index='Borough', aggfunc='sum')

# Display the pivot table
print("Pivot Table - Total Student Enrollment by Borough:")
print(pivot_table_enrollment)

# Find the borough with the highest total student enrollment
highest_enrollment_borough = pivot_table_enrollment.idxmax()

# Find the borough with the lowest total student enrollment
lowest_enrollment_borough = pivot_table_enrollment.idxmin()

# Display the results
print("Borough with the Highest Total Student Enrollment:", highest_enrollment_borough)
print("Borough with the Lowest Total Student Enrollment:", lowest_enrollment_borough)
```

Q14: Modify the pivot table from the step above to also reflect the average student body size (*Student Enrollment*) across boroughs, in the same pivot table.

```python
#Q14
# Pivot table to summarize total and average student enrollment across all 5 boroughs
pivot_table_enrollment = sat.pivot_table(values='Student Enrollment', index='Borough', aggfunc=['sum', 'mean'])

# Display the modified pivot table
print("Modified Pivot Table - Total and Average Student Enrollment by Borough:")
print(pivot_table_enrollment)
```

Q15: Create a pivot table of high schools from the Queens borough (*Borough*), containing the SAT section scores (*SAT Section*, *Score*) as columns and the school name (*School Name*) as index. Sort the table in descending order by math section scores.

```python
#Q15
# Assuming 'sat' is your dataframe
# Replace this with your actual dataframe name

# Filter data for high schools in Queens
queens_high_schools = sat[(sat['Borough'] == 'Queens') & (sat['School Name'].notnull())]

# Pivot table with SAT section scores as columns
pivot_table_queens = queens_high_schools.pivot_table(values='Score', index='School Name', columns='SAT Section', aggfunc='mean')

# Sort the table in descending order by math section scores
pivot_table_queens_sorted = pivot_table_queens.sort_values(by='Math', ascending=False)

# Display the sorted pivot table
print("Pivot Table of High Schools in Queens with SAT Section Scores:")
print(pivot_table_queens_sorted)
```
